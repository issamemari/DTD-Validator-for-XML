```python
import pandas

dataframe = pandas.read_csv("results1.csv", header=None)

dataframe[0] = dataframe[0] / (1024 * 1024)
dataframe[1] = dataframe[1] / 1000
dataframe[2] = dataframe[2] / (1024 * 1024)
```

```python
import matplotlib
import matplotlib.pylab as plt

plt.figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')

plt.plot(dataframe.values[0:, 0], dataframe.values[0:, 1], 'ro')
plt.xlabel('Document Size (Megabytes)')
plt.ylabel('Execution Time (Seconds)')
plt.show()
```

```python
import matplotlib
import matplotlib.pylab as plt


plt.figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')

plt.xlabel('Document Size (Megabytes)')
plt.ylabel('Memory Consumtion (Megabytes)')

plt.plot(dataframe.values[0:, 0], dataframe.values[0:, 2], 'ro')
plt.show()
```
